# Question no 2-How many runs were scored in each season?


```python
# Merge df1 and df2 to link runs to seasons
merged_df = pd.merge(df1, df2[['id', 'year']], on='id', how='left')

# Calculate total runs per season
runs_per_season = merged_df.groupby('year')['total_runs'].sum()

print("Total runs scored in each season:")
display(runs_per_season)
```

# Question no 8-Does winning the toss imply winning the game?



```python
# Count the number of matches where the toss winner is also the match winner
toss_and_match_wins = df2[df2['toss_winner'] == df2['winner']]
num_toss_and_match_wins = len(toss_and_match_wins)

# Total number of matches
total_matches = len(df2)

# Calculate the percentage
win_percentage_after_toss = (num_toss_and_match_wins / total_matches) * 100

print(f"Out of {total_matches} matches, the team that won the toss also won the match {num_toss_and_match_wins} times.")
print(f"This is a winning percentage of {win_percentage_after_toss:.2f}%.")
```

# Question no 32.Which team has scored the most runs in the first 6 overs?


```python
# Filter for deliveries in the first 6 overs (powerplay)
powerplay_deliveries = df1[df1['over'] <= 6]

# Group by batting team and sum the total runs in the powerplay
powerplay_runs = powerplay_deliveries.groupby('batting_team')['total_runs'].sum()

# Get the team with the most runs in the powerplay
most_powerplay_runs_team = powerplay_runs.idxmax()
most_powerplay_runs_count = powerplay_runs.max()

print(f"The team that has scored the most runs in the first 6 overs is: {most_powerplay_runs_team} with {most_powerplay_runs_count} runs.")
```

# Question no 33-Which team has scored the most runs in the last 4 overs?

```python
# Filter for deliveries in the last 4 overs (overs 17, 18, 19, 20)
last_4_overs_deliveries = df1[df1['over'] >= 17]

# Group by batting team and sum the total runs in the last 4 overs
last_4_overs_runs = last_4_overs_deliveries.groupby('batting_team')['total_runs'].sum()

# Get the team with the most runs in the last 4 overs
most_last_4_overs_runs_team = last_4_overs_runs.idxmax()
most_last_4_overs_runs_count = last_4_overs_runs.max()

print(f"The team that has scored the most runs in the last 4 overs is: {most_last_4_overs_runs_team} with {most_last_4_overs_runs_count} runs.")
```

#Question34:Which team has the best scoring run-rate in the first 6 overs?

```python
# Filter for deliveries in the first 6 overs (powerplay)
powerplay_deliveries = df1[df1['over'] <= 6].copy() # Use .copy() to avoid SettingWithCopyWarning

# Calculate total runs and total balls faced by each team in the powerplay
powerplay_team_runs = powerplay_deliveries.groupby('batting_team')['total_runs'].sum()
powerplay_team_balls = powerplay_deliveries.groupby('batting_team').size()

# Calculate run rate (runs per over) - 6 balls per over
# Add a small value to balls_faced to avoid division by zero
powerplay_run_rate = (powerplay_team_runs / (powerplay_team_balls / 6)).sort_values(ascending=False)


print("Run rate in the first 6 overs for each team:")
display(powerplay_run_rate)

# Get the team with the best run rate in the powerplay
best_powerplay_run_rate_team = powerplay_run_rate.idxmax()
best_powerplay_run_rate_value = powerplay_run_rate.max()

print(f"\nThe team with the best scoring run-rate in the first 6 overs is: {best_powerplay_run_rate_team} with a run rate of {best_powerplay_run_rate_value:.2f}.")
```
